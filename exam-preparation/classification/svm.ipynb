```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np

iris = load_iris()
X, y = iris["data"][:, (2, 3)], (iris["target"] == 2).astype(np.float64)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)

# Show classes
print("Features:", iris.feature_names)
print("Classes:", iris.target_names)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC, SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Linear
svm_classifier = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("linear_svc", LinearSVC(C=5, loss="hinge")),
    ]
)

# Polynomial
poly_svm_classifier = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("poly_svc", SVC(kernel="poly", degree=3, C=5, coef0=1)),
    ]
)

# RBF
# Gamma -> 'scale' / number
rbf_svm_classifier = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("rbf_svc", SVC(kernel="rbf", gamma="scale", C=5)),
    ]
)

svm_classifier.fit(X_train, y_train)
y_pred = svm_classifier.predict(X_test)


acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```
