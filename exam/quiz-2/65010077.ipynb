```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
t = np.linspace(0, 2 * np.pi, 1000) # radian

x = np.sin(t)
y = np.cos(t)

plt.figure(figsize=(6, 6))

plt.plot(x[x < 0], y[x < 0], color='green')

plt.plot(x[x >= 0], y[x >= 0], color='pink')

plt.title("2 separated circle lines")
plt.axis('equal')
plt.show()
```

```python
df = pd.read_csv("./netflix_titles.csv")
df
```

```python
missing = (
    df[
        [
            "show_id",
            "type",
            "title",
            "director",
            "cast",
            "country",
            "date_added",
            "release_year",
            "rating",
            "duration",
            "listed_in",
            "description",
        ]
    ]
    .isnull()
    .sum()
)
missing
```
